```python
from ultralytics import YOLO

# 1. โหลดโมเดลตั้งต้น (เราใช้ไซส์ n = nano เพราะเบาและเร็วที่สุด เหมาะกับสตรีมวิดีโอ)
# ถ้าอยากใช้ YOLOv8 ให้ใส่ "yolov8n.pt" / ถ้าอยากลองของใหม่ให้ใส่ "yolov11n.pt"
model = YOLO("yolo26n.pt") 

# 2. ตั้งค่าและเริ่มเทรนโมเดล
print("🚀 เริ่มกระบวนการเทรนโมเดล...")
results = model.train(
    data="my_dataset/data.yaml", # ชี้ไปที่ไฟล์ตั้งค่า Dataset ของเรา
    epochs=150,                   # จำนวนรอบที่ให้ AI เรียนรู้ (50 รอบกำลังสวยสำหรับเริ่มทดสอบ)
    imgsz=416,                   # ขนาดภาพมาตรฐานที่ YOLO ชอบ
    batch=8,                     # ถ้าคอมพิวเตอร์ RAM น้อย ให้ลดเหลือ 4 หรือ 2 
    device="0",                # ⚠️ ถ้าคุณมีการ์ดจอ NVIDIA ให้เปลี่ยนจาก "cpu" เป็น 0 (เลขศูนย์) จะเทรนเร็วขึ้น 10 เท่า!
    name="sign_language_model"   # ชื่อโฟลเดอร์ผลลัพธ์
)
print("🎉 เทรนเสร็จสมบูรณ์!")
```
